```python
import pandas as pd

RANDOM_SEED = 42
HOLDOUT_FRACTION = 0.30

data_dir = "data_for_bucket/input_data/togo/togo_full_2018_10"

subscribers = pd.read_csv(f"{data_dir}/combined_real_subscribers.csv")
print(f"Total subscribers: {len(subscribers)}")
subscribers.head()
```

```python
shuffled = subscribers.sample(frac=1, random_state=RANDOM_SEED).reset_index(drop=True)

n_holdout = int(len(shuffled) * HOLDOUT_FRACTION)
holdout = shuffled.iloc[:n_holdout]
train = shuffled.iloc[n_holdout:]

print(f"Holdout set: {len(holdout)} subscribers ({len(holdout)/len(subscribers):.1%})")
print(f"Training set: {len(train)} subscribers ({len(train)/len(subscribers):.1%})")
```

```python
out_path = f"{data_dir}/hold_out_subscribers.csv"
holdout.to_csv(out_path, index=False)
print(f"Saved holdout set to {out_path}")
```
