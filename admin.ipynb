# Admin Notebook

Utilities for managing the featurization evaluation service.
Requires `gcloud` CLI authenticated with a `.berkeley.edu` account.

```python
import subprocess
import requests

SERVICE_URL = 'https://featurization-test-server-594466465768.us-central1.run.app'

id_token = subprocess.run(
    ['gcloud', 'auth', 'print-identity-token'],
    capture_output=True, text=True, check=True
).stdout.strip()
```

## View all user counters for the current week

```python
response = requests.get(
    f"{SERVICE_URL}/get_counters",
    headers={"Authorization": f"Bearer {id_token}"},
    timeout=30,
)
print(f"Status: {response.status_code}")
data = response.json()
print(f"Week: {data['year']}-W{data['week']:02d}")
print("\nCounters:")
for user, count in sorted(data['counters'].items()):
    print(f"  {user}: {count}")
```

## Reset a user's counter

```python
response = requests.post(
    f"{SERVICE_URL}/reset_counter",
    headers={"Authorization": f"Bearer {id_token}"},
    json={"user": "selker@berkeley.edu"},
    timeout=30,
)
print(f"Status: {response.status_code}")
print(response.json())
```

## Health check 

```python
response = requests.get(
    f"{SERVICE_URL}/",
    headers={"Authorization": f"Bearer {id_token}"},
    timeout=30,
)
print(f"Status: {response.status_code}")
data = response.json()
print(data)
```
